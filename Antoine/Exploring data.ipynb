# Import

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

def file name & directory

```python
directory = "../../Challenge1-AM/DataBase/38Q31TzlO-"
file_name = "/npz_data/data.npz"
params_file_name = "/Minamo_Parameters-Wall2D.txt"
```

# Loading Data

```python
# read npz file
npz = np.load(directory+str(1)+file_name)
df = pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index').T
df['Sequence'] = 1 # add column Sequence

# read params file
with open(directory+str(1)+params_file_name) as params_file:
        lines = params_file.read().splitlines()
        break_time = lines[1].split(' = ')[1]
df['Break_time'] = break_time # add column Break_Time

for i in range(2,122):
    print(i)
    # read npz file
    npz = np.load(directory+str(i)+file_name)
    df2 = pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index').T
    df2['Sequence'] = i # add column Sequence
    
    # read params file
    with open(directory+str(i)+params_file_name) as params_file:
        lines = params_file.read().splitlines()
        break_time = lines[1].split(' = ')[1]
    df2['Break_time'] = break_time # add column Break_Time
    
    df = pd.concat([df, df2], axis=0)
```
