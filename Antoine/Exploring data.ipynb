# Import & loading Data

```python
npz = np.load(directory+str(1)+file_name)
df = pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index').T
df['Sequence'] = 1
for i in range(2,122):
    print(i)
    npz = np.load(directory+str(i)+file_name)
    df2 = pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index').T
    df2['Sequence'] = i
    df = pd.concat([df, df2], axis=0)
```
